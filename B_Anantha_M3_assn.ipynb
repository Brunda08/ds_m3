# 5. Data Preparation

## 5.1 1. Handling Missing Data

### Drop Columns & Imputing Missing Values

**DROP**

In the initial step of data preparation, we are addressing columns with a high percentage of missing values. Columns with more than 30% missing data may not provide meaningful insights and can be computationally expensive to handle. Thus, we will drop such columns to streamline our dataset.

```python
# Dropping columns with a high percentage of missing values
threshold = 0.3  # Adjusting the threshold as needed
wine_df = wine_df.dropna(thresh=len(wine_df) * (1 - threshold), axis=1)
```

After dropping columns with a high percentage of missing values, we observe that no essential columns have been removed. This action helps clean the dataset by removing attributes with substantial missing data, making it more manageable for subsequent analysis.

**IMPUTE**

In this step, we are addressing missing values in specific columns by imputing them with the mean of their respective columns. This imputation ensures that we retain valuable information while dealing with missing data.

```python
# Imputing missing values with the mean
wine_df['ResidualSugar'].fillna(wine_df['ResidualSugar'].mean(), inplace=True)
wine_df['Chlorides'].fillna(wine_df['Chlorides'].mean(), inplace=True)
wine_df['FreeSulfurDioxide'].fillna(wine_df['FreeSulfurDioxide'].mean(), inplace=True)
wine_df['TotalSulfurDioxide'].fillna(wine_df['TotalSulfurDioxide'].mean(), inplace=True)
wine_df['pH'].fillna(wine_df['pH'].mean(), inplace=True)
wine_df['Sulphates'].fillna(wine_df['Sulphates'].mean(), inplace=True)
wine_df['Alcohol'].fillna(wine_df['Alcohol'].mean(), inplace=True)
wine_df['STARS'].fillna(wine_df['STARS'].mean(), inplace=True)

```

By imputing missing values with the mean for the selected columns, we have retained valuable information in these attributes. This process ensures that we don't lose data while addressing missing values, improving the completeness of our dataset.

**CHECK**

In this step, we are verifying if any missing values still exist in the dataset after handling them through dropping and imputation.

```python
# Check for missing values
missing_values = wine_df[column_names].isnull().sum()
print("Missing Values:\n", missing_values)
```

The check for missing values reveals that there are no remaining missing values in the dataset. All essential attributes have been appropriately managed, ensuring the dataset is ready for further analysis and modeling. 

## 5.2 Data Integrity Issues

### 5.2.1 Handling Invalid Data Values

To ensure data quality, we are addressing rows with invalid 'pH' values in the dataset. We assume that pH values should fall within a specific valid range (typically between 0 and 14). Rows with 'pH' values outside this range will be removed from the dataset.

```python
# Removing rows with pH values outside the valid range (adjust the range as needed)
valid_pH_range = (0, 14)  # pH range from 0 to 14
wine_df = wine_df[(wine_df['pH'] >= valid_pH_range[0]) & (wine_df['pH'] <= valid_pH_range[1])]
```

By removing rows with 'pH' values outside the valid range of 0 to 14, we have ensured that the dataset only contains valid pH values. This step contributes to data quality and reliability, as it eliminates potentially erroneous data points that could affect subsequent analysis and modeling.

**CHECK**

After handling invalid 'pH' values and making necessary adjustments, we will perform a final check to ensure there are no missing values or outliers.



```python
# Check for missing values after handling invalid 'pH' values
missing_values = wine_df[column_names].isnull().sum()
print("Missing Values:\n", missing_values)
```

The final check for missing values indicates that there are no remaining missing values in the dataset after handling the invalid 'pH' values and addressing other missing data. The dataset is now prepared and free from any significant data integrity issues.

## 5.3 Data Distribution Issues

### 5.3.1 Addressing Skewness

In this step, we are addressing data distribution issues related to skewness in the 'pH' feature. We'll apply a log transformation to mitigate skewness, making the data distribution more symmetrical and conducive to better modeling.

```python
# Applying log transformation to reduce skewness in 'pH'
wine_df['pH'] = np.log1p(wine_df['pH'])
```

The log transformation has been applied to the 'pH' feature, effectively reducing its skewness. This transformation aligns the data distribution more closely to a normal distribution, enhancing the suitability of 'pH' for modeling purposes.

### 5.3.2 Feature Scaling

To ensure that all numerical features have similar scales, a crucial requirement for many machine learning algorithms, we are standardizing the numerical attributes. This process involves scaling the features to have a mean of 0 and a standard deviation of 1.

```python
# Standardizing numerical features using StandardScaler ( taken from GOOGLE)

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
numerical_columns = wine_df.select_dtypes(include=['number']).columns
wine_df[numerical_columns] = scaler.fit_transform(wine_df[numerical_columns])
```

By standardizing the numerical features, we have ensured that all attributes have similar scales. This standardization prepares the dataset for machine learning algorithms, as it eliminates potential biases arising from features with varying scales. The data is now ready for modeling, with improved compatibility for various algorithms and better model performance.

**CHECK**

After addressing skewness and performing feature scaling, we will perform a final check to confirm that the data distribution issues have been effectively resolved.



```python
# Check the data distribution after addressing skewness and feature scaling
data_distribution = wine_df.describe()
print("Data Distribution:\n", data_distribution)
```

The final data check confirms that the data distribution issues related to skewness have been successfully addressed through the log transformation of 'pH.' Additionally, feature scaling has been applied to ensure that all numerical features have similar scales. These data preparation steps have enhanced the dataset's suitability for machine learning, improving its overall quality and distribution.

## 5.4 Feature Engineering

### 5.4.1 Creating New Features

In the realm of feature engineering, we are creating a new feature called 'TotalAcidity' by summing the values of 'FixedAcidity' and 'VolatileAcidity.' This new feature captures the combined acidity of the wine, which can be a valuable attribute in our analysis.

```python
# Creating a new feature 'TotalAcidity' by summing 'FixedAcidity' and 'VolatileAcidity'
wine_df['TotalAcidity'] = wine_df['FixedAcidity'] + wine_df['VolatileAcidity']
```

The 'TotalAcidity' feature has been successfully created by summing 'FixedAcidity' and 'VolatileAcidity.' This new attribute provides a consolidated measure of acidity, potentially offering insights into the wine's characteristics that were not apparent with individual acidity components.

### 5.4.2 Binning or Categorization of 'pH'

To better understand the 'pH' attribute, we are categorizing it into pH levels, such as 'Acidic,' 'Neutral,' and 'Alkaline,' based on specified pH range thresholds. This categorization can simplify the analysis and interpretation of pH data.

```python
# Categorizing 'pH' into 'Acidic,' 'Neutral,' and 'Alkaline' based on pH range thresholds
pH_bins = [0, 6.5, 7.5, 14]  # Adjusting the pH range thresholds as needed
pH_labels = ['Acidic', 'Neutral', 'Alkaline']
wine_df['pH_Level'] = pd.cut(wine_df['pH'], bins=pH_bins, labels=pH_labels)
```

The 'pH' column has been effectively categorized into pH levels: 'Acidic,' 'Neutral,' and 'Alkaline,' based on the specified pH range thresholds. This categorization simplifies the interpretation of pH data, allowing for more straightforward analysis and insights into the acidity levels of the wines.
